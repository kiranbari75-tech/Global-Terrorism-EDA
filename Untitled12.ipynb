```python
# now we want to check top 10 country who effect terrism
country_attacks = df['country_txt'].value_counts()
country_attacks.head(10)
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
```

```python
# creating plt for to 10 country
top10 = country_attacks.head(10)
top10.plot(kind='bar')
plt.xlabel("Country")
plt.ylabel("Number of Attacks")
plt.title("Top 10 Countries with Highest Terrorist Attacks")
plt.show()
```

```python
# lets checks out how much attacks in india
india_df = df[df['country_txt'] == 'India']
```

```python
state_attacks = india_df['provstate'].value_counts()
state_attacks
```

```python
# lets see india to ten state attacks
top10_states = state_attacks.head(10)

top10_states.plot(kind='bar')
plt.xlabel("State")
plt.ylabel("Number of Attacks")
plt.title("Top 10 Indian States with Highest Terrorist Attacks")
plt.xticks(rotation=45)
plt.show()
```

```python
year_attacks = india_df['iyear'].value_counts().sort_index()
year_attacks
```

```python
plt.plot(year_attacks.index, year_attacks.values)
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.title("Year-wise Terrorist Attacks in India")
plt.show()
```

```python
# which city is most effected by terrer
city_attacks = india_df['city'].value_counts()
state_attacks = india_df['provstate'].value_counts()
city_attacks
state_attacks

```

```python
# creating heat map
city_df = india_df[india_df['city'] != 'Unknown']

top_cities = city_df['city'].value_counts().head(10).index
city_df = city_df[city_df['city'].isin(top_cities)]
```

```python
city_year = pd.pivot_table(
    city_df,
    index='city',
    columns='iyear',
    values='eventid',
    aggfunc='count',
    fill_value=0
)
```

```python
plt.figure(figsize=(14,6))
sns.heatmap(city_year, cmap="Reds")
plt.title("City vs Year Terrorist Attacks (India)")
plt.xlabel("Year")
plt.ylabel("City")
plt.show()
```

```python
# India has witnessed a significant number of fatalities due to terrorist
#incidents, with deaths concentrated in specific high-conflict states and
#peak years.
total_killed = india_df['nkill'].fillna(0).sum()
print("Total people killed in India:", int(total_killed))
```

```python
region_attacks = df['region_txt'].value_counts()
print(region_attacks)
region_year = df.groupby(['region_txt', 'iyear']).size().unstack(fill_value=0)
```

```python
top_regions = region_attacks.head(5).index
region_year_top = region_year.loc[top_regions]

plt.figure(figsize=(12,5))
for region in region_year_top.index:
    plt.plot(region_year_top.columns, region_year_top.loc[region], label=region)

plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.title("Region-wise Yearly Terrorist Attacks (Top 5 Regions)")
plt.legend()
plt.show()
```

```python
# now we want to check type of attack
attack_types = df['attacktype1_txt'].value_counts()
print(attack_types)
```

```python
# now we attack type of india
india_df = df[df['country_txt'] == 'India']

india_attack_types = india_df['attacktype1_txt'].value_counts()
print(india_attack_types)
```

```python
top_types = india_attack_types.head(3).index

attack_year = india_df[india_df['attacktype1_txt'].isin(top_types)] \
                .groupby(['iyear', 'attacktype1_txt']).size().unstack(fill_value=0)

attack_year.plot(figsize=(12,5))
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.title("Year-wise Top Attack Types in India")
plt.show()
```
